```python
import matplotlib.pyplot as plt
import numpy as np
import random

times = [100*i for i in range(1,11)]
ans = 2022
```

```python
def s1(n):
    res = []
    for i in range(n):
        num = 1
        choice = random.randrange(0, 10000)
    
        while(choice != ans):
            choice = random.randrange(0, 10000)
            num+=1
        res.append(num)
    return res
    
```

```python
trial1_sd = []
trial1_mean = []
for t in times:
    trial = s1(t)
    trial1_sd.append(np.std(trial))
    trial1_mean.append(np.mean(trial))
```

```python
trial1_sd.__len__()
```

```python
def s2(n):
    res = []
    for i in range(n):
        set = {}
        choice = random.randrange(0,10000)
        set[choice] = 1
        num = 1
        
        while(choice != 2022):
            choice = random.randrange(0,10000)
            while(set.get(choice) != None):
                choice = random.randrange(0,10000)
            set[choice] = 1
            num+=1
        res.append(num)
    return res
s2(3)
```

```python
trial2_sd = []
trial2_mean = []
for t in times:
    trial = s2(t)
    trial2_sd.append(np.std(trial))
    trial2_mean.append(np.mean(trial))
```

```python
labels = times
x_pos = np.arange(len(labels))


fig,(ax1,ax2) = plt.subplots(1,2,figsize=(12, 4))
ax1.bar(x_pos, trial1_mean,
       yerr=trial1_sd,
       align='center',
       alpha=0.5,
       ecolor='black',
       capsize=10)
ax1.set_ylabel('Average number of trials')
ax1.set_xticks(x_pos)
ax1.set_xticklabels(labels)
ax1.set_title('Strategy 1')
ax1.yaxis.grid(True)


ax2.bar(x_pos, trial2_mean,
       yerr=trial2_sd,
       align='center',
       alpha=0.5,
       ecolor='black',
       capsize=10)
ax2.set_ylabel('Average number of trials')
ax2.set_xticks(x_pos)
ax2.set_xticklabels(labels)
ax2.set_title('Strategy 2')
ax2.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()
plt.savefig('bar_plot_with_error_bars.png')
plt.show()
```
